```python
data = df.drop(columns=['Education','City','Gender','EverBenched'])
```

```python

for column in data.columns:
    plt.figure(figsize=(6, 4))
    sns.boxplot(x=data[column])
    plt.title(f'Ящик с усами для {column}')
    plt.show()

'Построение гистограмм для каждого столбца'
for column in data.columns:
    plt.figure(figsize=(6, 4))
    sns.histplot(data[column],bins=20, kde=True)
    plt.title(f'Гистограмма для {column}')
    plt.show()
```

```python
num_rows = len(df.index)
low_information_cols = []

for col in df.columns:
    value_counts = df[col].value_counts(dropna=False)
    top_pct = (value_counts / num_rows).iloc[0]

    if top_pct > 0.80:
        low_information_cols.append(col)
        print(f'Столбец "{col}" имеет {top_pct*100:.2f}% одинаковых значений.')
```

```python
# Комбинации признаков
feature_combinations = list(itertools.combinations(data.columns, 2))

for feat1, feat2 in feature_combinations:
    sns.jointplot(x=feat1,
                  y=feat2,
                  data=df,
                  palette='viridis')

plt.show()
```
